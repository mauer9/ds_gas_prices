```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('data-source/gas_prices.csv')
gas
```

```python
plt.figure(figsize=(8,4))
plt.title('Gas price per country')

plt.plot(gas.Year, gas['South Korea'], '.-k', label='South Korea')
plt.plot(gas.Year, gas['Australia'], '.-c', label='Australia')
plt.plot(gas.Year, gas['Canada'], '.-r', label='Canada')
plt.plot(gas.Year, gas['USA'], '.-b', label='United States')

plt.xticks(gas.loc[::3, 'Year'])
plt.xlabel('Years')
plt.ylabel('Gas price / USD')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Gas price per country')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.loc[::3, 'Year'])
plt.xlabel('Years')
plt.ylabel('Gas price / USD')

plt.legend()
plt.show()
```
